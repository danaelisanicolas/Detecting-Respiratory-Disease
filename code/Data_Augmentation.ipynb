```python
def count_files(loc):
    file_count = []
    
    for cat in cats:
        path, dirs, files = next(os.walk(loc + cat))
        count = len(files)
        file_count.append({'diagnosis': cat, 'count' : count})
    
    return file_count
```

```python
def is_wav(filename):
    '''
        Checks if files are .wav files
        Utility tool in converting wav to png files
    '''
    return filename.split('.')[-1] == 'wav'
```

```python
def manipulate_audio(data, sampling_rate, shift_max=0, shift_direction='none', speed_factor=0, is_shift=False):
    '''
    Augment sound file using either shifting time or speed.
    Required input: audio data and sampling rate
    Returns the new augmented data
    '''
    if is_shift == True:
        shift = np.random.randint(sampling_rate * shift_max)
        if shift_direction == 'right':
            shift = -shift
        elif shift_direction == 'both':
            direction = np.random.randint(0, 2)
            if direction == 1:
                shift = -shift
        
        augmented_data = np.roll(data, shift)
        # Set to silence for heading/ tailing
        if shift > 0:
            augmented_data[:shift] = 0
        else:
            augmented_data[shift:] = 0
        return augmented_data
    else:
        return lb.effects.time_stretch(data, speed_factor)
```

```python
def manipulate_for_n_times(location='../', n=1, verbose=False):
    '''
    Input Location = directory of the audio files that will be resampled
    New augmented audio will be saved in the same directory
    '''    
    for i in range(n):
        #randomise any audio file in the directory
        audio_file = ''
        while (True): 
            audio_file = random.choice(os.listdir(location))
            if ('augmented' in audio_file) == True:
                #dont augment augmented file
                continue
                
            if is_wav(audio_file) == True:
                break
        
        if verbose == True:
            print('augmenting: ' + audio_file + '...')
        
        #read audio file
        data_24, samplerate_24 = sf.read(location + audio_file)

        will_shift = float(np.random.randint(2))        
        shift_or_speed_factor = np.random.uniform(0.1, 1.0)
        
        vanilla_audio_file = audio_file.split('.wav')[0]

        ts = datetime.datetime.now().timestamp() 

        if will_shift == True:
            direction = 'right'
            shift_dir = float(np.random.randint(2))
            if shift_dir == 1.0:
                direction = 'both'
                
            if verbose == True:
                print('augmenting by shifting...' + direction)
                
            new_audio = manipulate_audio(data_24, sampling_rate=samplerate_24, 
                                         shift_max=shift_or_speed_factor, 
                                         shift_direction=direction, is_shift=will_shift)
            
            #convert to wav file and save on same directory
            sf.write(location + vanilla_audio_file + '_augmented_' + str(ts) + '.wav', data = new_audio, 
                    samplerate = samplerate_24)
        else:
            if verbose == True:
                print('augmenting by changing speed by...' + str(shift_or_speed_factor))

            new_audio = manipulate_audio(data_24, samplerate_24, speed_factor=shift_or_speed_factor)

            #convert to wav file and save on same directory
            sf.write(location + vanilla_audio_file + '_augmented_' + str(ts) + '.wav', data = new_audio, 
                    samplerate = samplerate_24)
```

```python
#set limit to 600 files per folder
main_loc = '../data/Respiratory_Files/New_Audio/train/'
for cat in cats:
    print('-'*150)
    print('working on ' + cat)
    print('-'*150)
    
    path, dirs, files = next(os.walk(main_loc + cat))
    if len(files) <= 0 or cat == 'COPD':
        continue
        
    #work only on directories with at least 1 file
    n = 600 - len(files)
    if n > 0:
        #run only if it's less than 600 files
        manipulate_for_n_times(main_loc + cat + '/', n=n, verbose=True)
```
